```python
from sankeyflow import Sankey
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 10), dpi=144)
nodes = [
    [('Applications', 84, {'color': 'gray'})],
    [('Replies', 21, {'color': 'green'}), ('Rejections (1)', 41, {'color': 'red'}), ('No replies', 22, {'color': 'orange'})],
    [('Phone Screening', 11, {'color': 'green'}), ('Online Assessment (1)', 2, {'color': 'green'}), ('Online Interview (1)', 2, {'color': 'green'}), ('On-site Interview (1)', 1, {'color': 'green'}), ('Rejected by me (1)', 5, {'color': 'green'})],
    [('On-site Interview (2)', 5, {'color': 'green'}), ('Online Assessment (2)', 3, {'color': 'green'}), ('Rejected by me (2)', 4, {'color': 'green'}), ('Rejections (2)', 4, {'color': 'red'})],
    [('Online Interview (2)', 1, {'color': 'green'}), ('Rejected by me (3)', 2, {'color': 'green'}), ('Rejections (3)', 4, {'color': 'red'}), ('Ghosted', 1, {'color': 'orange'})],
    [('Offer', 1, {'color': 'green'})],
]

flows = [
    ('Applications', 'Replies', 21),
    ('Applications', 'Rejections (1)', 41),
    ('Applications', 'No replies', 22),
    ('Replies', 'Phone Screening', 11),
    ('Replies', 'Online Assessment (1)', 2),
    ('Replies', 'Rejected by me (1)', 5),
    ('Replies', 'Online Interview (1)', 2),
    ('Replies', 'On-site Interview (1)', 1),
    ('Phone Screening', 'Rejections (2)', 3),
    ('Phone Screening', 'On-site Interview (2)', 4),
    ('Phone Screening', 'Online Assessment (2)', 3),
    ('Phone Screening', 'Rejected by me (2)', 1),
    ('Online Assessment (1)', 'Rejections (2)', 1),
    ('Online Assessment (1)', 'Rejected by me (2)', 1),
    ('Online Interview (1)', 'Rejected by me (2)', 2),
    ('On-site Interview (1)', 'On-site Interview (2)', 1),
    ('On-site Interview (2)', 'Rejections (3)', 2),
    ('On-site Interview (2)', 'Online Interview (2)', 1),
    ('On-site Interview (2)', 'Rejected by me (3)', 2),
    ('Online Assessment (2)', 'Ghosted', 1),
    ('Online Assessment (2)', 'Rejections (3)', 2),
    ('Online Interview (2)', 'Offer', 1),


]
s = Sankey(
    flows=flows,
    nodes=nodes,
)

s.find_node('Rejections (1)')[0].label='Rejections'
s.find_node('Rejections (2)')[0].label='Rejections'
s.find_node('Rejections (3)')[0].label='Rejections'
s.find_node('Rejected by me (1)')[0].label='Rejected by me'
s.find_node('Rejected by me (2)')[0].label='Rejected by me'
s.find_node('Rejected by me (3)')[0].label='Rejected by me'
s.find_node('On-site Interview (1)')[0].label='On-site Interview'
s.find_node('On-site Interview (2)')[0].label='On-site Interview'
s.find_node('Online Interview (1)')[0].label='Online Interview'
s.find_node('Online Interview (2)')[0].label='Online Interview'
s.find_node('Online Assessment (1)')[0].label='Online Assessment'
s.find_node('Online Assessment (2)')[0].label='Online Assessment'
s.draw()
plt.title('Job Applications Januar - April 2025', fontsize=20)
plt.show()
plt.savefig('../plots/job_applications.png', bbox_inches='tight')
```
